# Dropping columns

```python
df.drop(columns = ["Unnamed: 0", "Photo", "Flag", "Club Logo", "Special", "International Reputation", "Skill Moves", "Real Face", "Unnamed: 89"], inplace = True)
df.drop(df.loc[:, 'LS':'RB'], inplace = True, axis = 1)
```

# Cleaning the data

### Seeing what's missing

```python
df.isnull().sum()
```

### Dropping the rows which don't have clubs

```python
df[df.Club.isnull()]
```

```python
df.dropna(subset = ['Club'], inplace = True)
```

```python
df.shape
```

### Dropping the rows which don't specify preffered foot, and other details

```python
df[df['Preferred Foot'].isnull()]
```

```python
df.dropna(subset = ['Preferred Foot'], inplace = True)
```

```python
df.shape
```

### Changing "Release Clause" to float64 from string

```python
start, stop, step = 1, -1, 1
df["Release Clause"] = df["Release Clause"].str.slice(start, stop, step)
df["Release Clause"] = pd.to_numeric(df["Release Clause"])
```

```python
df["Value"] = df["Value"].str.slice(start, stop, step)
df["Value"] = pd.to_numeric(df["Value"], errors = "coerce")
```

```python
df["Wage"] = df["Wage"].str.slice(start, stop, step)
df["Wage"] = pd.to_numeric(df["Wage"], errors = "coerce")
```

```python
df["Crossing"] = df["Crossing"].str.slice(0, 2, 1)
```

```python
def pie():
    count = filterdf.Nationality.value_counts()
    
    fig1, ax1 = plt.subplots()
    ax1.pie(count, labels = count.index, autopct = '%1.f%%', startangle = 90, pctdistance = 0.6)
    
    #Drawing a circle
    centre_circle = plt.Circle((0, 0), 0.70, fc = 'white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    
    ax1.axis('equal')
    plt.tight_layout()
    plt.show()

def dpie():
    #Data to plot
    nat = filterdf.Nationality.value_counts()
    club = filterdf.Club.value_counts()
    
    #Plot
    #Outer circle
    plt.pie(nat, labels = nat.index, startangle=90)
    #Inner circle
    #plt.pie(club, radius=0.75, startangle = 90)
    plt.pie(club, labels = club.index, radius=0.75, startangle = 90, pctdistance = 0)
    centre_circle = plt.Circle((0, 0), 0.5, color = 'black', fc = 'white', linewidth = 0)
    fig = plt.gcf() #Get reference to current figure
    fig.gca().add_artist(centre_circle)
    
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

```python
#pie chart on Nationality if more filters are given
if ("Nationality" not in filters and "Club" not in filters):
    dpie()
elif ("Nationality" not in filters):
    pie()
```
